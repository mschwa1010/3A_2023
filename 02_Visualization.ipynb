```python
## Import libraries
## Until WS05
import numpy as np
import pandas as pd
import datetime

## Visualization
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

## For saving files
from pathlib import Path 
import pickle
```

## Trips Dataset

```python
## Load trips file
df_trips = pickle.load(open("trips.pkl","rb"))
df_trips.head()
```

```python
## Sort by date
df_sorted = df_trips.sort_values(by='Trip Start Timestamp') 
## Add unique index 
df_sorted['index'] = range(0, len(df_sorted))
df_sorted = df_sorted.set_index('index')
df_sorted.head()
```

Let us visualize some aspects of the dataframe.

```python
# Create a dictionary to map day of the month numbers to month names
month_dict = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July',
             8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

# Count the number of trips per day of the week
counts = df_trips['Trip Start Timestamp'].groupby(df_trips['Trip Start Timestamp'].dt.month).count()

# Map day of the week numbers to day of the week names and order by day, starting from Monday
counts.index = counts.index.map(month_dict)
counts = counts[['January', 'February', 'March', 'April','May',  'June',  'July',
              'August', 'September', 'October', 'November',  'December']]

# Print the counts
print(counts)
```

```python
import matplotlib.ticker as mtick
# Count the number of rows per month
df_counts = df_trips.groupby(pd.Grouper(key='Trip Start Timestamp', freq='M')).size().reset_index(name='count')

# Create the bar plot
ax = df_counts.plot(kind='bar', x='Trip Start Timestamp', y='count', color='darkblue', legend=False)

# Set the x-axis label and title
ax.set_xlabel('Month')
ax.set_ylabel('Number of Trips')
ax.set_title('Number of Trips per Month')

# Set the x-tick labels to month names
months = df_counts['Trip Start Timestamp'].dt.month_name().tolist()
ax.set_xticklabels(months)

# Format the y-axis tick labels to remove the e^ notation
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, pos: f'{x:,.0f}'))

```

```python
# Create a dictionary to map day of the week numbers to day of the week names
day_dict = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}

# Count the number of trips per day of the week
counts = df_trips['Trip Start Timestamp'].groupby(df_trips['Trip Start Timestamp'].dt.dayofweek).count()

# Map day of the week numbers to day of the week names and order by day, starting from Monday
counts.index = counts.index.map(day_dict)
counts = counts[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]

# Print the counts
print(counts)
```

```python
# Count the number of rows per day of the week
df_counts = df_trips.groupby(pd.Grouper(key='Trip Start Timestamp', freq='D')).size().reset_index(name='count')
df_counts['day_of_week'] = df_counts['Trip Start Timestamp'].dt.day_name()
df_counts['day_of_week'] = pd.Categorical(df_counts['day_of_week'], categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)

# Create the bar plot
ax = df_counts.groupby('day_of_week').sum().plot(kind='bar', y='count', color='darkblue', legend=False)

# Set the x-axis label and title
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Number of Trips')
ax.set_title('Number of Trips per Day of the Week')

# Set the x-tick labels to day of the week names
ax.set_xticklabels(df_counts['day_of_week'].cat.categories)

# Format the y-axis tick labels to remove the e^ notation
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, pos: f'{x:,.0f}'))

# Show the plot
plt.show()
```

```python
df_trips['Trip Start Timestamp'].groupby(df_trips['Trip Start Timestamp'].dt.hour).agg('count')
```

```python
# Convert the timestamp column to datetime format
df_trips['Trip Start Timestamp'] = pd.to_datetime(df_trips['Trip Start Timestamp'])

# Group the data by hour of the day and count the number of rows for each hour
df_counts = df_trips.groupby(df_trips['Trip Start Timestamp'].dt.hour).size().reset_index(name='count')

# Create the bar plot
ax = df_counts.plot(kind='bar', x='Trip Start Timestamp', y='count', color='darkblue', legend=False)

# Set the x-axis label and title
ax.set_xlabel('Hour of the Day')
ax.set_ylabel('Number of Trips')
ax.set_title('Number of Trips per Hour of the Day')

# Format the y-axis tick labels to remove the e^ notation
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, pos: f'{x:,.0f}'))
```
